```python
# === SETUP — run this first ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings; warnings.filterwarnings('ignore')

# downloading essential libraries

url = 'https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv'
df = pd.read_csv(url)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce').fillna(0)
print('Loaded:', df.shape)

# printing the shape of the dataframe that we have downlaoded
```

```python
df.head()
# first 5 or 10 rows
```

```python
df.shape
# shape of the dataframe
```

```python
df.dtypes
# data types of the columns
```

```python
df['Churn'].value_counts()
# count of the variable in column "Churn"
```

```python
long_term = df[df['tenure'] > 50]
print('Long-term customers:', len(long_term))
long_term.head()

# viewing the first 5 or 10 rows whose tenure is greater than 50
```

```python
top5 = df.sort_values('TotalCharges', ascending=False).head(5)
top5[['customerID', 'tenure', 'MonthlyCharges', 'TotalCharges']]

# viewing the top 5 customers with the highest total charges and their customer ID, tenure, monthly charges, and total charges
```

```python
df.groupby('Contract')['MonthlyCharges'].mean().round(2)

# grouping the data by the contract and calculating the mean of the monthly charges
```

```python
df.groupby('Churn').agg({'MonthlyCharges': 'mean', 'tenure': 'mean'}).round(2)

# grouping the data by the churn and calculating the mean of both monthly charges and tenure
```

```python
pd.crosstab(df['SeniorCitizen'], df['Churn'])

# creating a crosstab of senior citizen and churn
```
